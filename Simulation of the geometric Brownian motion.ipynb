# Importing some basic libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Defining the parameters, $Sigma$ is the volatility

```python
mu = 0.1
n = 100
T = 1
M = 100
S0 = 100
sigma = 0.3
```

# Simulation of the geometric Brownian motion

```python
dt = T/n
St = np.exp((mu-sigma**2/2)*dt+sigma*np.transpose(np.random.normal(0,np.sqrt(dt), size = (M,n))))

St = np.vstack([np.ones(M), St])

St = S0*St.cumprod(axis = 0)
```

# Viewing one of the stacks

```python
St[1]
```

Time intervals in years

```python
time = np.linspace(0,T,n+1)
array = np.transpose(np.full((M,n+1), fill_value = time))
```

```python
plt.plot(array, St)
plt.xlabel("Years $(t)$")
plt.ylabel("Stock Price $(S_t)$")
plt.title("realization of the geometric Brownian motion \n $dS_t = \mu S_tdt +\sigma S_t dW_t$")
plt.show()
```
